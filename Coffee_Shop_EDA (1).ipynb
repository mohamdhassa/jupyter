```python
# Adjusting data types
#--> Convert transaction_date to datetime
df_coffee_sales['transaction_date'] = pd.to_datetime(df_coffee_sales['transaction_date'])
df_coffee_sales['customer_since'] = pd.to_datetime(df_coffee_sales['customer_since'])
df_coffee_sales.dtypes
```

```python
df_coffee_sales['transaction_datetime'] = df_coffee_sales['transaction_date'].astype("str")+' '+df_coffee_sales['transaction_time']
df_coffee_sales.head()
```

```python
df_coffee_sales['transaction_datetime'] = pd.to_datetime(df_coffee_sales['transaction_datetime'])
df_coffee_sales.dtypes
```

```python
# Convert customer_id to str
df_coffee_sales['customer_id'] = df_coffee_sales['customer_id'].astype('str')
df_coffee_sales.dtypes
```

```python
# Convert customer_birth_year to int
df_coffee_sales['customer_birth_year'].isna().sum() #-> Count the null values
# Replace null values in customer_birth_year with 0
df_coffee_sales['customer_birth_year'] = df_coffee_sales['customer_birth_year'].fillna(0)
df_coffee_sales['customer_birth_year'].isna().sum() #-> Count the null values
# Convert customer_birth_year to int
df_coffee_sales['customer_birth_year'] = df_coffee_sales['customer_birth_year'].astype('int')
df_coffee_sales.dtypes
```

```python
# Change the column names:
# --> 'line_item_amount' to 'retail price'
df_coffee_sales = df_coffee_sales.rename(columns={'line_item_amount':'retail price'})
df_coffee_sales.head()
```

```python
df_coffee_sales.transaction_datetime
df_coffee_sales['retail price']
```

```python
# Create a New column to calculate:
# --> 'total_wholesale_price' = current_wholesale_price * quantity
df_coffee_sales['total_wholesale_price'] = df_coffee_sales['current_wholesale_price'] * df_coffee_sales['quantity']
df_coffee_sales.head()
# --> 'profit' = retail price - current_wholesale_price
df_coffee_sales['profit'] = df_coffee_sales['retail price'] - df_coffee_sales['current_wholesale_price']
df_coffee_sales
```
