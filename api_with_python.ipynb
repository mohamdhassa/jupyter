The OMDb API is a RESTful web service to obtain movie information, all content and images on the site are contributed and maintained by our users.

```python
API_KEY = '169932e1'   #--> Always keep it stored somewhere!
# This is just for reference 
# url = f'http://www.omdbapi.com/?apikey={API_KEY}&t={MOVIE_NAME}&type=movie'
# omdb_req = requests.get(url)
# omdb_data = omdb_req.json()
```

```python
# Read movie names from imdb_movie_names.csv
df_movie_name = pd.read_csv('data/imdb_movie_names.csv')
df_movie_name.head()
```

#### One Way of Requesting the data:
Requesting the data for each movie (10 times) and each time we add it to a list 

```python
# Run the request on 10 movies - You can try to do more later on
# Looping over a dataframe
movie_details = []
for name in df_movie_name['Movie Names'].to_list()[:10]:
    url = f'http://www.omdbapi.com/?apikey={API_KEY}&t={name}\
    &type=movie'
    omdb_req = requests.get(url)
    omdb_data = omdb_req.json()
    movie_details.append(omdb_data)
movie_details
```

```python
# Keep the following columns. Note you might need to do extra work for the some columns
# Title, Type, Year, Released, Rated, Runtime, Genre, Metascore, imdbRating, imdbVotes, BoxOffice, Internet Movie Database, Rotten Tomatoes, Metacritic
df_movies = pd.DataFrame(movie_details)
df_movies.head()
```

```python
# Clean up the data
# Title, Type, Year, Released, Rated, Runtime, Genre, Metascore, imdbRating, imdbVotes, BoxOffice, Internet Movie Database, Rotten Tomatoes, Metacritic
# Note : The columns Internet Movie Database, Rotten Tomatoes, and Metacritic and inside the Ratings column
df_movies['Ratings'].iloc[0]

# Instead of having them all in a column, we can add each in a new column
df_movies['Internet Movie Database'] = df_movies['Ratings'].iloc[0][0]['Value']
df_movies['Rotten Tomatoes'] = df_movies['Ratings'].iloc[0][1]['Value']
df_movies['Metacritic'] = df_movies['Ratings'].iloc[0][2]['Value']

# Select the columns that we need
df_movies = df_movies[['Title', 'Type', 'Year', 'Released', 'Rated', 'Runtime', 
                       'Genre', 'Metascore', 'imdbRating', 'imdbVotes', 'BoxOffice', 
                       'Internet Movie Database', 'Rotten Tomatoes', 'Metacritic']]
df_movies.head()
```

#### Another way to Request the data
Requesting the data and only taking the information needed and adding them to a list of their own

```python
# Run the request on 10 movies - You can try to do more later on
# Looping over a dataframe
movie_details_ver2 = {
    'Title':[], 
    'Type':[], 
    'Year':[], 
    'Released':[], 
    'Rated':[], 
    'Runtime':[], 
    'Genre':[], 
    'Metascore':[], 
    'imdbRating':[], 
    'imdbVotes':[], 
    'BoxOffice':[],
    'Internet Movie Database':[], 
    'Rotten Tomatoes':[], 
    'Metacritic':[]
}
for name in df_movie_name['Movie Names'].to_list()[:10]:
    url = f'http://www.omdbapi.com/?apikey={API_KEY}&t={name}&type=movie'
    omdb_req = requests.get(url)
    omdb_data = omdb_req.json()
    movie_details_ver2['Title'].append(omdb_data['Title'])
    movie_details_ver2['Type'].append(omdb_data['Type'])
    movie_details_ver2['Year'].append(omdb_data['Year'])
    movie_details_ver2['Released'].append(omdb_data['Released'])
    movie_details_ver2['Rated'].append(omdb_data['Rated'])
    movie_details_ver2['Runtime'].append(omdb_data['Runtime'])
    movie_details_ver2['Genre'].append(omdb_data['Genre'])
    movie_details_ver2['Metascore'].append(omdb_data['Metascore'])
    movie_details_ver2['imdbRating'].append(omdb_data['imdbRating'])
    movie_details_ver2['imdbVotes'].append(omdb_data['imdbVotes'])
    movie_details_ver2['BoxOffice'].append(omdb_data['BoxOffice'])
    movie_details_ver2['Internet Movie Database'].append(omdb_data['Ratings'][0]['Value'])
    movie_details_ver2['Rotten Tomatoes'].append(omdb_data['Ratings'][1]['Value'])
    movie_details_ver2['Metacritic'].append(omdb_data['Ratings'][2]['Value'])
movie_details_ver2
```
